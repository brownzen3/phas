Plot the following fourth-order Legendre polynomial:

$$
P_4(x) = \frac{1}{8} \left( 35 x^4 - 30 x^2 + 3 \right)
$$

(over the range $-1<x<1$)

```python
import numpy as np
import matplotlib.pyplot as plt

# magic command for interactive features
#%matplotlib ipympl

def P4(x):
    """
    Fourth-order Legendre polynomial: takes one input, x, and outputs the value of the polynomial at x.
    """
    return((1/8) * ( 35*(x**4) - 30*(x**2) + 3 ))

x = np.linspace(-1,1, 1000) # plot P4(x) over the range -1 < x < 1
y = P4(x)

# plot P4(x)
plt.figure()
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("P₄(x)")
plt.title("A plot of P₄(x)")
plt.grid(which='major', axis='both', color='#cccccc') # enable gridlines
```
